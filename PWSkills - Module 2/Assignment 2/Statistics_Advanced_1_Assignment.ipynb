**7. What is the significance of confidence intervals in statistical analysis?**

- Definition:

 - A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence.

 - For example, a 95% confidence interval for the mean might be:

 - CI=[48.3,52.7]

 - Which means: "We are 95% confident that the true population mean lies between 48.3 and 52.7."

- üìå Why Confidence Intervals Are Important:

| --------------------------------- | ---------------------------------------------------------------------------------------- |
|  Reflects **uncertainty**        | Unlike point estimates, CI gives a **range** to express uncertainty in sample statistics |
|  Enables **inference**           | Helps estimate **unknown population parameters** based on sample data                    |
|  Aids **decision making**        | Businesses, scientists, and policymakers use CIs to assess **risk and reliability**      |
|  Avoids **false certainty**      | Provides a balanced view instead of misleading single-value estimates                    |

**9. Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.**

- Here is a complete Python program that:

 - Generates 1000 random numbers from a normal distribution with
mean = 50 and standard deviation = 5

 - Computes the sample mean and standard deviation using NumPy

 - Draws a histogram to visualize the distribution using matplotlib

- What We‚Äôll See:

 - A bell-shaped histogram centered around 50

 - Red dashed line showing the sample mean

 - Approximate spread of data from 45 to 55 (due to std = 5)

- Python Program:

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from normal distribution
mean = 50
std_dev = 5
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Step 2: Compute mean and standard deviation using NumPy
computed_mean = np.mean(data)
computed_std = np.std(data)

print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std:.2f}")

# Step 3: Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.axvline(computed_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {computed_mean:.2f}')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Step 1: Calculate sample mean and standard deviation
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample standard deviation
n = len(daily_sales)

# Step 2: Calculate 95% Confidence Interval using CLT
z = 1.96  # Z-score for 95% confidence
margin_of_error = z * (std_sales / np.sqrt(n))

lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

# Step 3: Display result
print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
