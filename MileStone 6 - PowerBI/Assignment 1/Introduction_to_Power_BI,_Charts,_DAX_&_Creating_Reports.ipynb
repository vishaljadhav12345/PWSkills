**Question 1 :**

**Define Power BI and What are the key components of the Power BI ecosystem? Briefly
explain:**

**‚óè Power BI Desktop**

**‚óè Power BI Service**

**‚óè Power BI Mobile**

**‚óè Power BI Gateway**

**Answer :**

Definition of Power BI

- Power BI is a business analytics and data visualization tool developed by Microsoft.

- It allows users to connect to different data sources, transform data, create interactive reports and dashboards, and share insights across an organization.

- Power BI helps convert raw data into meaningful visual insights for better decision-making.

‚úÖ Key Components of the Power BI Ecosystem

1Ô∏è‚É£ Power BI Desktop

- A Windows application used to build data models, reports, and visualizations.

- Offers data transformation using Power Query and data modeling using DAX (Data Analysis Expressions).

- Used mainly by data analysts to create complete report files (.pbix).

- Supports connections to databases, Excel files, cloud services, and more.

2Ô∏è‚É£ Power BI Service (PowerBI.com)

- A cloud-based platform where reports and dashboards created in Power BI Desktop are published.

- Allows sharing, collaboration, scheduled refresh, and dashboard creation.

- Enables organizations to view and interact with reports online.

- Also supports creating workspaces for team collaboration.

3Ô∏è‚É£ Power BI Mobile

- A mobile application available for Android, iOS, and Windows devices.

- Allows users to view and interact with dashboards and reports on the go.

- Supports mobile-optimized layouts created in Power BI Desktop.

- Useful for managers and executives to access insights anytime, anywhere.

4Ô∏è‚É£ Power BI Gateway

- A bridge that connects on-premises data sources (like SQL Server, Oracle, local files) with the Power BI Service.

- Ensures secure data transfer for scheduled refreshes without uploading data manually.

- Two types:

   - Personal Gateway: For individual use

   - Enterprise Gateway: For organizational use, supports multiple users & sources

- Helps keep cloud reports updated with real-time on-premise data.

‚≠ê Summary Table

| Component            | Purpose                                        |
| -------------------- | ---------------------------------------------- |
| **Power BI Desktop** | Build reports, transform data, create models   |
| **Power BI Service** | Publish, share, collaborate, online dashboards |
| **Power BI Mobile**  | View dashboards on mobile devices              |
| **Power BI Gateway** | Connect on-premises data to cloud reports      |


**Question 2 :**

**Compare the following Power BI visuals:**

**‚óè Pie Chart vs Donut Chart**

**‚óè Bar Chart vs Column Chart**

**When would you prefer one over the other? Give one example for each pair.**

**Answer :**

1Ô∏è‚É£ Pie Chart vs Donut Chart

- A Pie Chart is a circular visual used in Power BI to display percentage contribution of each category to a whole.

- Each slice represents a category, and the size of the slice is proportional to its value.

- Pie charts are best used when you have limited categories (preferably 3‚Äì5) and when each category is visually distinct.

- They are simple and widely used, but can become cluttered if too many slices exist.

- A Donut Chart, on the other hand, is a slightly modified version of a pie chart.

- It contains a central hole, which makes the design cleaner and more modern.

- Donut charts allow additional context to be displayed in the center, such as total sales, total profit, or any KPI value.

- The readability of a donut chart is often considered better than a pie chart because the empty center focuses the viewer's attention on the relative proportions.

- When to Use Pie Chart

   - A pie chart is useful when you want a quick understanding of the distribution among a small number of categories.

- Example:

  - Showing Sales Contribution of Four Regions (North, South, East, West) as percentages.

- When to Use Donut Chart

   - A donut chart is preferred when you want a cleaner look, show totals at the center, or compare more than four categories.

- Example:

  - Displaying Profit Share by Product Category with Total Profit shown at the center.

2Ô∏è‚É£ Bar Chart vs Column Chart

- A Bar Chart represents data using horizontal bars. It is extremely useful when category names are long, or when the number of categories is large.

- Since horizontal space is usually more than vertical space, bar charts are excellent for displaying many categories at once.

- It is one of the most readable visuals for making direct comparisons between items.

- A Column Chart represents data using vertical bars. Column charts are ideal when comparing values across time periods, such as years, months, or quarters.

- Human eyes naturally follow vertical height to compare values, making column charts intuitive for trend-based analysis.

- Column charts work best when the number of categories is smaller and follow a natural sequence.

- When to Use Bar Chart

  - Use a bar chart when you have many categories or long text labels.

- Example:

  - Comparing Sales across 12 Product Sub-Categories like ‚ÄúOffice Supplies ‚Äì Stationery‚Äù or ‚ÄúFurniture ‚Äì Chairs‚Äù.

- When to Use Column Chart

  - Use a column chart when dealing with time-series data or sequential data.

- Example:

- Monthly Sales Trends for a year (Jan‚ÄìDec) to analyze seasonality.

‚≠ê Summary

| Visual           | Key Feature              | Best Use Case                              | Example                                  |
| ---------------- | ------------------------ | ------------------------------------------ | ---------------------------------------- |
| **Pie Chart**    | Circular, no center hole | Few categories, simple proportion analysis | Sales share by region                    |
| **Donut Chart**  | Has center hole          | Showing totals + cleaner look              | Profit share with total profit in center |
| **Bar Chart**    | Horizontal bars          | Many categories, long labels               | Sales by sub-category                    |
| **Column Chart** | Vertical bars            | Time trends, sequential comparison         | Monthly sales trend                      |


**Question 3 :**

**Explain the significance of:**

**‚óè Star schema vs Snowflake schema**

**‚óè Primary key vs Foreign key in relationships (Power BI)**

**Why is cardinality important?**

**Answer :**

1Ô∏è‚É£ Star Schema vs Snowflake Schema

‚úî Star Schema

- A Star Schema is a simple and denormalized data modeling approach where one central Fact Table is connected to multiple Dimension Tables.

- It gets its name because the diagram resembles a star.

- Features:

  1. Very few relationships

  2. Fast query performance

  3. Dimensions are denormalized (all attributes stored in one table)

  4. Easy to understand for analysts and Power BI users

- Advantages:

  1. Best performance for reporting tools like Power BI

  2. Simplifies DAX calculations

  3. Ideal for dashboards and analytics

  4. Reduces joins ‚Üí faster refresh and visuals

- Example:

  - FactSales connected to DimCustomer, DimProduct, DimDate, DimRegion.

‚úî Snowflake Schema

- A Snowflake Schema is a more normalized version of the Star Schema where dimension tables are further split into sub-dimensions.

- Features:

  1. Dimensions are normalized into multiple related tables

  2. More tables ‚Üí more joins

  3. Reduces data redundancy

- Advantages:

  1. Saves storage

  2. Better data integrity

  3. Useful for large and complex databases

- When to Use:

  1. When data duplication must be avoided

  2. For enterprise-level warehouse systems

‚≠ê Key Difference Summary

| Feature    | Star Schema         | Snowflake Schema             |
| ---------- | ------------------- | ---------------------------- |
| Structure  | Simple              | Complex                      |
| Dimensions | Denormalized        | Normalized                   |
| Speed      | Faster              | Slower                       |
| Storage    | More                | Less                         |
| Best For   | Power BI dashboards | Large, normalized warehouses |


2Ô∏è‚É£ Primary Key vs Foreign Key in Power BI Relationships

‚úî Primary Key

- A Primary Key is a column (or set of columns) in a table that uniquely identifies each row.

- Power BI uses primary keys in dimension tables.

- Example:

  - CustomerID in DimCustomer
DateKey in DimDate

- Significance:

  - Ensures no duplicate values

  - Forms the ‚Äúone‚Äù side of a relationship

‚úî Foreign Key

- A Foreign Key is a column in a table that refers to the Primary Key of another table.
Used in fact tables.

- Example:

  - CustomerID in FactSales is a foreign key referencing DimCustomer.

- Significance:

  - Connects facts with dimensions

  - Forms the ‚Äúmany‚Äù side of a relationship

‚≠ê Relationship Example in Power BI

- DimCustomer.CustomerID (Primary Key) ‚Üí FactSales.CustomerID (Foreign Key)

- This enables Power BI to filter facts based on customer attributes.

3Ô∏è‚É£ Why is Cardinality Important?

- Cardinality defines the type of relationship between tables in Power BI:

1. One-to-Many (1:*): Most common, best performance

2. Many-to-One (*:1): Reverse direction

3. Many-to-Many (:): Used only when necessary

4. One-to-One (1:1): Rare

- Why Cardinality Matters

1. Controls how filters flow from one table to another

2. Improves model performance by reducing unnecessary relationships

3. Ensures accurate calculations in DAX

4. Helps avoid incorrect aggregations or duplicate rows

5. Ensures the data model follows best practices (star schema)

**Question 4 :**

**Differentiate between:**

**‚óè Calculated column vs Measure Also, define Row context and Filter context with simple examples.**

**Answer :**

1Ô∏è‚É£ Calculated Column vs Measure

‚úî Calculated Column

- A calculated column is a new column created in a table using DAX.

- It is computed row-by-row during data refresh and stored in the data model.

- Characteristics:

  - Computed when the data model loads

  - Stored physically in the table (takes memory)

  - Has row context

  - Useful for labeling, categories, or filtering visuals

- Example:

  - Profit = Sales[SalesAmount] - Sales[Cost]


  - A Profit column is added for each row.

‚úî Measure

- A measure is a calculation performed on the fly, based on filters applied in a report.
Measures are not stored; they are computed dynamically.

- Characteristics:

  - Calculated at query time

  - Does not consume data model space

  - Has filter context

  - Used for aggregations (SUM, AVG, COUNT, etc.)

- Example:

  - Total Sales = SUM(Sales[SalesAmount])


- The value changes depending on slicers, filters, and visuals.

‚≠ê Key Differences

| Feature     | Calculated Column       | Measure                      |
| ----------- | ----------------------- | ---------------------------- |
| Computation | Static (during refresh) | Dynamic (during report view) |
| Storage     | Stored in model         | Not stored                   |
| Context     | Row context             | Filter context               |
| Use Case    | Row-level logic         | Aggregations, KPIs           |
| Performance | Slower, uses memory     | Faster, efficient            |


2Ô∏è‚É£ Row Context

- Row Context means Power BI evaluates a DAX expression one row at a time.

- It exists in:

  - Calculated Columns

  - Iterator functions (SUMX, AVERAGEX)

- Simple Example:

- If you create a calculated column:

  - TotalCost = Sales[Quantity] * Sales[UnitPrice]


- Power BI calculates this for each row, independently.

- Row context knows which row it is currently calculating.

3Ô∏è‚É£ Filter Context

- Filter Context refers to filters applied to a calculation through:

  - Slicers

  - Visual filters

  - Row headers

  - CALCULATE()

- Simple Example:

- A measure:

  - Total Sales = SUM(Sales[SalesAmount])


- If you select India in a slicer, the measure recalculates only India‚Äôs sales.

- This happens because filter context modifies the calculation.

‚≠ê Easy Summary

1. Calculated Column:
Row-by-row, stored in the model, uses row context.

2. Measure:
Calculated dynamically, used for aggregations, uses filter context.

3. Row Context:
Evaluates expression for each row independently.

4. Filter Context:
Filters applied from visuals, slicers, or CALCULATE().

**Question 5:**

**What is the difference between a report and a dashboard in Power BI?**

**Answer :**

- Power BI provides two key ways to present insights‚ÄîReports and Dashboards‚Äîbut they serve different purposes and have different capabilities.

- Understanding these differences is essential for effective data analysis and visualization.

1Ô∏è‚É£ Power BI Report (Detailed Multi-Page View)

- A report is a multi-page, interactive document created using Power BI Desktop or Power BI Service.

- Reports allow users to explore data in depth through multiple visuals, filters, drill-downs, and detailed analysis.

- Characteristics of Reports:

1. Can have multiple pages

2. Highly interactive with features like slicers, bookmarks, drill-through

3. Created in Power BI Desktop (.pbix file)

4. Shows visuals from a single dataset

5. Supports data modeling, DAX, and transformations

6. Best for analysts and developers who need detailed analysis

- Example:

1. A Sales Report with separate pages for

2. Sales Summary

3. Profit Analysis

4. Customer Insights

5. Region-wise Performance

2Ô∏è‚É£ Power BI Dashboard (Single-Page Summary View)

- A dashboard is a single-page, high-level view created in Power BI Service.

- It presents key business metrics through ‚Äútiles‚Äù pinned from one or more reports or datasets.

- Characteristics of Dashboards:

1. One page only

2. Created in Power BI Service, not Desktop

3. Tiles can come from multiple reports and datasets

4. Used for high-level KPIs and real-time monitoring

5. Ideal for managers and executives

6. Not as interactive as reports (no drill-down within tiles)

- Example:

- A Business Dashboard showing:

1. Total Sales

2. Profit Margin

3. Top 5 Products

4. Daily Sales Trend

5. Target vs Achievement

6. All in a single page.

3Ô∏è‚É£ Key Differences Between Report and Dashboard (Tabular Form)

| Feature           | Report                        | Dashboard                    |
| ----------------- | ----------------------------- | ---------------------------- |
| **Pages**         | Multiple                      | Single page                  |
| **Creation Tool** | Power BI Desktop/Service      | Power BI Service only        |
| **Data Sources**  | One dataset                   | Multiple datasets            |
| **Interactivity** | High (drill-through, slicers) | Limited                      |
| **Purpose**       | Detailed analysis             | Quick summary/KPI monitoring |
| **Sharing**       | Through workspaces            | Sharable, mobile-friendly    |
| **File Type**     | .pbix                         | No file‚Äîonline view          |


4Ô∏è‚É£ When to Use What?

- Use a Report When:

1. You need detailed exploration and analysis

2. Multiple pages are required

3. You want interactive visuals

4. You want to apply filters, slicers, and drill-downs

- Use a Dashboard When:

1. You need a high-level overview

2. You want to combine data from multiple sources

3. You want a quick, mobile-friendly KPI view

4. You need real-time monitoring
