**1. What is hypothesis testing in statistics?**

- Hypothesis testing is a formal statistical method used to make inferences or draw conclusions about a population based on sample data. It allows us to test an assumption (called a hypothesis) about a population parameter.

- Key Concepts:

| Term                                  | Description                                                                                        |
| ------------------------------------- | -------------------------------------------------------------------------------------------------- |
| **Null Hypothesis (H‚ÇÄ)**              | A statement that there is no effect or no difference. It is the default or status quo assumption.  |
| **Alternative Hypothesis (H‚ÇÅ or Ha)** | A statement that contradicts the null hypothesis; it suggests a new effect or difference exists.   |
| **Test Statistic**                    | A numerical value calculated from the sample data, used to decide whether to reject H‚ÇÄ.            |
| **p-value**                           | The probability of obtaining a test statistic as extreme as the one observed, assuming H‚ÇÄ is true. |
| **Significance Level (Œ±)**            | A threshold (commonly 0.05 or 5%) for rejecting H‚ÇÄ. If p-value ‚â§ Œ±, we reject H‚ÇÄ.                  |
| **Type I Error**                      | Rejecting H‚ÇÄ when it is actually true.                                                             |
| **Type II Error**                     | Failing to reject H‚ÇÄ when it is actually false.                                                    |

- Steps in Hypothesis Testing:

1. Formulate Hypotheses

 - H‚ÇÄ: Œº = 50

 - H‚ÇÅ: Œº ‚â† 50

2. Select Significance Level (Œ±)

 - Common values: 0.05, 0.01, 0.10

3. Collect Sample Data

4. Compute Test Statistic

 - Depends on the test: Z-test, T-test, Chi-square test, etc.

5. Make Decision

 - If p-value ‚â§ Œ±: Reject H‚ÇÄ

 - If p-value > Œ±: Fail to reject H‚ÇÄ

 - Example:

   Suppose a company claims their battery lasts 10 hours on average.

 - H‚ÇÄ: Œº = 10

 - H‚ÇÅ: Œº ‚â† 10

   You test 30 batteries and find the average life is 9.5 hours. A hypothesis test can determine if this difference is statistically significant.

- Why Hypothesis Testing is Important?

 - Validates business decisions based on data

 - Used in quality control, medicine, psychology, and economics

 - Helps in comparing products, treatments, policies, and more

**2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?**

1. Null Hypothesis (H‚ÇÄ):

 - The null hypothesis is a default or starting assumption in hypothesis testing.

 - It states that there is no effect, no difference, or that the observed result is due to random chance.

 - It is the hypothesis that is tested directly and possibly rejected.

- Example:

 - If a drug company claims their medicine has no effect on blood pressure:

 - H‚ÇÄ: The drug has no effect on blood pressure. (Œº = Œº‚ÇÄ)

2. Alternative Hypothesis (H‚ÇÅ or Ha):

 - The alternative hypothesis is a contradictory statement to the null.

 - It represents what we are trying to prove ‚Äî that there is an effect, or a significant difference.

 - If the evidence is strong enough (p-value ‚â§ Œ±), we reject H‚ÇÄ in favor of H‚ÇÅ.

- Example:

 - H‚ÇÅ: The drug has an effect on blood pressure. (Œº ‚â† Œº‚ÇÄ)

- Comparison Table:

| Aspect                 | Null Hypothesis (H‚ÇÄ)                | Alternative Hypothesis (H‚ÇÅ)              |
| ---------------------- | ----------------------------------- | ---------------------------------------- |
| Meaning                | No change, no effect, no difference | There is a change, effect, or difference |
| Goal                   | Try to disprove or reject it        | Try to support it                        |
| Outcome if p-value ‚â§ Œ± | Reject H‚ÇÄ                           | Accept H‚ÇÅ                                |
| Default assumption     | Always assumed to be true initially | Only accepted if evidence supports it    |

- Visual Representation:

 Suppose you're testing whether a new teaching method is better than the traditional one:

 - H‚ÇÄ: New method = Traditional method

 - H‚ÇÅ: New method ‚â† Traditional method

 If you get strong evidence, you reject H‚ÇÄ and accept that the new method may be better (or worse).

**3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

- What is Significance Level (Œ±)?

  The significance level, denoted by Œ± (alpha), is the threshold probability used in hypothesis testing to determine whether to reject the null hypothesis (H‚ÇÄ).

 - It represents the maximum acceptable risk of making a Type I error (i.e., rejecting a true null hypothesis).

 - Commonly used values:

   - Œ± = 0.05 (5%)

   - Œ± = 0.01 (1%)

   - Œ± = 0.10 (10%)

- Role of Significance Level in Hypothesis Testing

 - After calculating the p-value from the test statistic, it is compared with Œ±.

 - Decision Rule:

   - If p-value ‚â§ Œ± ‚Üí Reject H‚ÇÄ ‚Üí The result is statistically significant.

   - If p-value > Œ± ‚Üí Fail to Reject H‚ÇÄ ‚Üí The result is not statistically significant.

- Example:

 Let‚Äôs say you are testing if a new fertilizer increases crop yield:

 - H‚ÇÄ: The fertilizer has no effect.

 - H‚ÇÅ: The fertilizer increases yield.

 - You set Œ± = 0.05

 If your statistical test gives:

 - p-value = 0.03 ‚Üí 0.03 < 0.05 ‚Üí Reject H‚ÇÄ

 - So, you conclude the fertilizer likely increases yield.

- Why is Significance Level Important?

| Feature                    | Explanation                                                                                    |
| -------------------------- | ---------------------------------------------------------------------------------------------- |
| **Controls Error Risk**    | Œ± defines how much error (Type I) you're willing to tolerate                                   |
| **Sets Decision Boundary** | Provides a clear rule to **accept or reject** H‚ÇÄ                                               |
| **Affects Confidence**     | A smaller Œ± increases **confidence** in rejecting H‚ÇÄ but may miss true effects (Type II error) |

- Common Misconception:

 - Œ± is not the probability that H‚ÇÄ is true.

 - It is the probability of rejecting H‚ÇÄ when it is actually true.

- Summary:

 - Significance Level (Œ±) is a pre-defined cutoff used to make objective decisions in hypothesis testing.

 - It helps us balance risk and confidence in statistical conclusions.

**4. What are Type I and Type II errors? Give examples of each.**

- In hypothesis testing, errors can occur when making decisions about the null hypothesis (H‚ÇÄ). These errors are categorized into:

- Type I Error (False Positive)

 Definition:

 - Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

 In simple terms:

 - You think there is an effect or difference when there isn't.

 Probability of Type I Error = Œ± (Significance level)

- Example of Type I Error:

 Imagine a COVID-19 test:

 - H‚ÇÄ: The person does not have COVID-19.

 - H‚ÇÅ: The person has COVID-19.

 If the test falsely detects COVID-19 (positive result when they don't have it) ‚Üí Type I Error.

 You're rejecting H‚ÇÄ (person doesn‚Äôt have COVID) even though it‚Äôs true.

- Type II Error (False Negative)

 Definition:

 - Failing to reject the null hypothesis (H‚ÇÄ) when it is actually false.

 In simple terms:

 - You miss an actual effect or difference.

 Probability of Type II Error = Œ≤

- Example of Type II Error:

 Same COVID-19 test:

 - H‚ÇÄ: The person does not have COVID-19.

 - H‚ÇÅ: The person has COVID-19.

 If the test misses the virus and gives a negative result for a truly infected person ‚Üí Type II Error.

 You're accepting H‚ÇÄ (no COVID) when it's actually false.

- Summary Table:

| Error Type        | What Happens                           | Decision       | Truth              | Risk Symbol |
| ----------------- | -------------------------------------- | -------------- | ------------------ | ----------- |
| **Type I Error**  | Reject H‚ÇÄ when it is **true**          | False positive | No real effect     | Œ± (alpha)   |
| **Type II Error** | Fail to reject H‚ÇÄ when it is **false** | False negative | Real effect exists | Œ≤ (beta)    |

| Scenario                  | Type of Error     |
| ------------------------- | ----------------- |
| Innocent person convicted | **Type I Error**  |
| Guilty person set free    | **Type II Error** |



**5. What is the difference between a Z-test and a T-test? Explain when to use
each.**

- Z-Test vs T-Test ‚Äî Overview:

| Feature                           | **Z-Test**                                | **T-Test**                               |
| --------------------------------- | ----------------------------------------- | ---------------------------------------- |
| Sample size                       | Large (n > 30)                            | Small (n ‚â§ 30)                           |
| Population standard deviation (œÉ) | **Known**                                 | **Unknown**                              |
| Distribution used                 | **Standard normal distribution (Z)**      | **Student's t-distribution**             |
| Shape of distribution             | Symmetrical, narrower tails               | Symmetrical, **fatter tails**            |
| Use Case                          | Testing population mean, proportion, etc. | Comparing means of one/two small samples |

- What is a Z-Test?

 Use When:

 - Sample size is large (typically > 30).

 - Population standard deviation (œÉ) is known.

 Types:

 - One-sample Z-test (compare sample mean to population mean)

 - Two-sample Z-test (compare means of two large samples)

 - Z-test for proportions

 Example:

 - You have test scores of 1000 students and you know the population standard deviation.

 - You want to test if the class average differs from the national average ‚Üí Use Z-test.

- What is a T-Test?

  Use When:

 - Sample size is small (typically ‚â§ 30).

 - Population standard deviation (œÉ) is unknown.

 Types:

 - One-sample t-test ‚Äì compare sample mean to a known value.

 - Two-sample t-test (independent) ‚Äì compare means of two independent groups.

 - Paired sample t-test (dependent) ‚Äì compare means from the same group at different times (before & after).

 Example:

 - You collect blood pressure readings from 15 patients before and after a drug treatment.

 - To check if the drug had a significant effect ‚Üí Use paired t-test.

- Summary Table:

| Criteria                 | **Z-Test**                      | **T-Test**                     |
| ------------------------ | ------------------------------- | ------------------------------ |
| Sample Size              | > 30                            | ‚â§ 30                           |
| Population œÉ known?      | Yes                             | No                             |
| Uses which distribution? | Normal (Z)                      | Student‚Äôs t-distribution       |
| Variance estimation      | From population                 | From sample                    |
| Example Use Case         | Large dataset, population known | Small experiments, A/B testing |

- Visualization:

 - T-distribution has fatter tails ‚Üí More uncertainty.

 - As sample size increases, t-distribution approaches Z-distribution.

- When to Use:

| Situation                                | Use          |
| ---------------------------------------- | ------------ |
| n > 30 and œÉ is known                    | Z-Test       |
| n ‚â§ 30 or œÉ is unknown                   | T-Test       |
| Comparing sample mean to population mean | Z or T       |
| Comparing two groups (A/B testing)       | Two-sample T |


**6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.**

- Here‚Äôs a Python program that:

  Generates random numbers following a binomial distribution with:

 - n = 10 (number of trials),

 - p = 0.5 (probability of success),

  Plots a histogram of the results using Matplotlib.

- Output:

 - A histogram showing the frequency of number of successes (0 through 10).

 - The mean and standard deviation printed for the distribution.

- Quick Recap of Binomial Distribution:

 - Discrete distribution.

 - Models the number of successes in n independent Bernoulli trials.

 - Each trial has only two outcomes: success (p) or failure (1-p).  

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10          # number of trials
p = 0.5         # probability of success
size = 1000     # number of random samples

# Generate binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Calculate statistics
mean = np.mean(binomial_data)
std_dev = np.std(binomial_data)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', align='left')
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

**7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.**

**sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

- Here's a complete solution to perform hypothesis testing using Z-statistics on the given sample data:

- Problem Setup:

 - You are testing whether the population mean is equal to a given value using the Z-test.

- Hypotheses:

  Let‚Äôs assume:

 - Null Hypothesis ùêª
0 : Œº = 50 (the population mean is 50)

 - Alternative Hypothesis
ùêª
1: Œº ‚â† 50 (the population mean is not 50)

 - Significance Level: Œ± = 0.05 (95% confidence)

- Conclusion:

 There is no statistically significant difference between the sample mean and the population mean of 50, at the 5% level of significance.

```python
import numpy as np
from scipy import stats

# Given data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to NumPy array
sample_data = np.array(sample_data)

# Sample size
n = len(sample_data)

# Sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Hypothesized population mean
mu = 50

# Z statistic
z_score = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.6f}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

**8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

- Here's a complete Python script to:

 1. Simulate data from a normal distribution

 2. Calculate the 95% confidence interval for the mean

 3. Plot the data using matplotlib

- Output Explanation

 - Mean of the sample is around 100

 - The 95% confidence interval tells us where the true population mean likely lies

 - The histogram visualizes the distribution with the mean and CI lines

- Python Script:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Generate random normal data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=100, scale=15, size=1000)  # mean=100, std=15

# Step 2: Calculate sample mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)  # standard error of the mean

# Step 3: Calculate 95% Confidence Interval
confidence = 0.95
h = std_err * stats.t.ppf((1 + confidence) / 2.0, len(data) - 1)
lower_bound = mean - h
upper_bound = mean + h

# Step 4: Print results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Step 5: Plot histogram with confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(lower_bound, color='red', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.axvline(mean, color='blue', linestyle='solid', linewidth=2, label='Mean')
plt.title('Histogram of Normally Distributed Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

- Here's a full answer for Question 9, including:

 - A Python function to calculate Z-scores

 - A histogram plot to visualize the standardized data

 - A clear explanation of what Z-scores represent

- Explanation: What Z-scores Represent

  A Z-score (also called a standard score) tells you how many standard deviations a data point is from the mean of the dataset.

  Interpretation:

  - Z = 0: Data point is exactly at the mean

 - Z > 0: Data point is above the mean

 - Z < 0: Data point is below the mean

 - Z = 1: 1 standard deviation above the mean

 - Z = -2: 2 standard deviations below the mean

 Use Cases:

 - Identify outliers (Z-scores beyond ¬±2 or ¬±3)

 - Compare values across different distributions

 - Normalize data for machine learning algorithms

- Python Code: Calculate Z-scores & Plot

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Function to calculate Z-scores and plot histogram of standardized data.

    Parameters:
        data (list or np.array): The input data

    Returns:
        np.array: Z-scores of the input data
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot histogram of Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
    plt.axvline(0, color='blue', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.axvline(1, color='green', linestyle='dashed', label='1 Std Dev')
    plt.axvline(-1, color='green', linestyle='dashed')
    plt.title("Histogram of Z-Scores")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=70, scale=10, size=1000)
z_scores = calculate_z_scores(sample_data)
```
